# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func, desc
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results
```

```python
# Code to find out the name and data type of the "measurement" data
inspector = inspect(engine)
columns_measurement = inspector.get_columns('measurement')
for c in columns_measurement:
    print(c['name'], c["type"])
```

```python
# Code to find out the name and data type of the "station" data
inspector = inspect(engine)
columns_station = inspector.get_columns('station')
for d in columns_station:
    print(d['name'], d["type"])
```

```python
# Calculate the date 1 year ago from the last data point in the database

# Calculating the last date of measurement in the data set
for row in session.query(Measurement.date, Measurement.prcp)\
.order_by(Measurement.date.desc()).limit(1):
    print(row)
```

```python
# Perform a query to retrieve the data and precipitation scores
year_pcrp = session.query(Measurement.date, Measurement.prcp)\
.filter(Measurement.date >= '2016-08-23').filter(Measurement.date <='2017-08-23').all()
```

```python
year_pcrp
```

```python
# Save the query results as a Pandas DataFrame and set the index to the date column
prcp_analysis = pd.DataFrame(year_pcrp, columns = ['Date', ' Precipitation'])
prcp_analysis['Date'] = pd.to_datetime(prcp_analysis['Date'])
prcp_analysis = prcp_analysis.set_index(['Date'])
prcp_analysis.head(5)
```

```python
# Use Pandas Plotting with Matplotlib to plot the data
#line plot
prcp_analysis.plot(sort_columns=True, rot=45, use_index=True, color = 'blue',\
                  figsize = (16,10), grid = True)
               
#labels
plt.title("Precipitation in Hawaii from 2016-08-23 until 2017-08-23")
plt.ylabel("Precipitation (in inches)")
plt.xlabel("Date")

plt.ylim(-0.5,7)
plt.show()
```

```python
# Use Pandas to calcualte the summary statistics for the precipitation data
prcp_analysis.describe()
```

# Station Analysis

```python
# Design a query to show how many stations are available in this dataset?
station_count = session.query(Station.name).count()
print(f"There is a total of {station_count} stations in the dataset")
```

```python
# What are the most active stations? (i.e. what stations have the most rows)?
# List the stations and the counts in descending order.
most_active = session.query(Measurement.station, Station.name, func.count(Measurement.tobs))\
.filter(Measurement.station == Station.station)\
.group_by(Measurement.station)\
.order_by(func.count(Measurement.tobs).desc()).all()
```

```python
most_active
```

```python
print(f"The most active station is {most_active[0]}")
```

```python
# Using the station id from the previous query, calculate the lowest temperature recorded, 
# highest temperature recorded, and average temperature of the most active station?

station_stats = session.query(func.min(Measurement.tobs),func.max(Measurement.tobs)\
,func.avg(Measurement.tobs)).filter(Measurement.station == 'USC00519281').all()

print(f"The mininmun, maximum and average temperatures recorded by\
station USC00519281 are {station_stats} respectively")
```

```python
# Choose the station with the highest number of temperature observations.
# Query the last 12 months of temperature observation data for this station and plot the results as a histogram

# Query to get the last year of tobs
year_tobs = session.query(Measurement.tobs)\
.filter(Measurement.station == 'USC00519281')\
.filter(Measurement.date >= '2016-08-23').filter(Measurement.date <='2017-08-23').all()
```

```python
# Verify results of query
print(year_tobs)
```

```python
# Place query into a list
tobs_list = [tobs[0] for tobs in year_tobs]
```

```python
# Histogram plot of results
import matplotlib.patches as mpatches
x = tobs_list
num_bins = 12
n, bins, patches = plt.hist(x, num_bins, facecolor='blue', alpha=0.8)

blue_label = mpatches.Patch(color='blue', label='tobs')
plt.legend(handles=[blue_label])
plt.ylabel('Frequency')
plt.title('Histogram of Most Common Temperature Observations in \n WAIHEE 837.5, HI US from 2016-08-23 until 2017-08-23')

plt.show()
```

## Bonus Section

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """
    
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```

```python
end_date = '2017-07-30'
```

```python
def calc_temps(start_date, end_date):
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date).filter(Measurement.date <= end_date).all()
print(calc_temps(start_date, end_date))
```

```python
# Using function `calc_temps` to calculate the tmin, tavg, and tmax 
# for your trip using the previous year's data for those same dates.
start_date2 = '2016-07-20'
end_date2 = '2016-07-30'
def calc_temps(start_date2, end_date2):
    return session.query(func.min(Measurement.tobs), func.avg(Measurement.tobs), func.max(Measurement.tobs)).\
        filter(Measurement.date >= start_date2).filter(Measurement.date <= end_date2).all()
print(calc_temps(start_date2, end_date2))
```

```python
# Placing values into dataframe
trip_temp = pd.DataFrame(calc_temps(start_date2, end_date2), columns = ['min', 'avg', 'max'])
trip_temp
```

```python
# Plot the results from your previous query as a bar chart. 
# Use "Trip Avg Temp" as your Title
# Use the average temperature for the y value
# Use the peak-to-peak (tmax-tmin) value as the y error bar (yerr)

error_bar = trip_temp['max']-trip_temp['min']

plt.bar(1, trip_temp['avg'], yerr=error_bar, align='center', color='blue', alpha=0.7, width=0.1)

plt.title('Trip Avg Temp')
plt.ylabel('Temperature (F)')
plt.ylim(0,100)

plt.tick_params(bottom=False, labelbottom=False)

plt.show()
```
